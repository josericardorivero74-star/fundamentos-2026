### Ejemplos prácticos para Ingeniería Agroindustrial

Aquí tienes tres scripts sencillos que ilustran tareas comunes en el área:

#### 1. Cálculo del Índice de Madurez (Relación Sólidos/Acidez)

Este código ayuda a determinar si una fruta está lista para la cosecha basándose en grados Brix y acidez titulable.

```
# Datos de entrada
grados_brix = 12.5
acidez_titulable = 0.8

# Cálculo del índice
indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez:.2f}")

if indice_madurez >= 15:
    print("Estado: Óptimo para cosecha/procesamiento.")
else:
    print("Estado: Aún le falta maduración.")
```

#### 2. Estimación de Merma en el Secado

Un cálculo rápido para saber cuánta masa se pierde al deshidratar un producto agrícola.

```
peso_inicial = 500  # kg
humedad_inicial = 0.80  # 80%
humedad_final = 0.12    # 12%

# Fórmula de balance de masa (sólidos constantes)
peso_final = peso_inicial * (1 - humedad_inicial) / (1 - humedad_final)

print(f"Tras el secado, los {peso_inicial}kg se convertirán en {peso_final:.2f}kg.")
print(f"Pérdida de agua (merma): {peso_inicial - peso_final:.2f}kg.")
```

#### 3. Clasificación Automática de Calidad

Un pequeño sistema lógico que clasifica un lote de producto según su peso.

```
pesos_lote = [150, 210, 185, 230, 140]  # gramos

for peso in pesos_lote:
    if peso < 150:
        categoria = "Segunda (Venta local)"
    elif 150 <= peso <= 200:
        categoria = "Primera (Nacional)"
    else:
        categoria = "Premium (Exportación)"

    print(f"Fruto de {peso}g -> Categoría: {categoria}")
```
