#**Ejercicio 1: Capacidad de Silos de Granos**

**Planteamiento**:

Se busca automatizar el cálculo de la capacidad de almacenamiento de un silo cilíndrico para maíz. Dado que el maíz ocupa un **volumen específico** y tiene una **densidad estándar**, el programa debe convertir dimensiones físicas (radio y altura) en masa (toneladas).

**Requerimientos del Problema**:

**Entradas**: Radio (r) en metros y Altura (h) en metros.

**Constantes**: Densidad del maíz ≈ 0.75 ton/m³.

**Fórmulas**:

**Volumen**: V = π × r² × h
Masa: M = V × 0.75

**Salida**:
Capacidad total en toneladas.

**Prueba del Programa**:

**Entrada**: Radio = 3 m, Altura = 10 m.


**Cálculo**:
V = π × 3² × 10 ≈ 282.74 m³
M = 282.74 × 0.75 ≈ 212.06 toneladas

**Salida**:
≈ 212.06 toneladas



```python
import math

#Entrada de datos
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

#Procesamiento
volumen = math.pi * (radio ** 2) * altura  # float
masa = volumen * 0.75 # float

#Salida de datos
print("La capacidad del silo es de:", masa, "toneladas")  # str y float
```

#**Ejercicio 2**: Dosificación de Fertilizante Líquido

**Planteamiento**:

En un sistema de riego por goteo, es vital calcular con precisión cuánto insumo comprar y cuánta agua se necesita para la mezcla, evitando el desperdicio o la sub-dosificación según el área de cultivo.

**Requerimientos del Problema**:

**Entradas**: Hectáreas totales y Dosis requerida (L/ha).

**Lógica de Dilución**: Relación 1:500 (por cada litro de fertilizante, se usan 500 litros de agua).

**Cálculos**:

**Producto Total** = Hectáreas × Dosis
Agua Total = Producto Total × 500

**Salida**:
Litros de producto puro y litros de agua necesarios.

**Prueba del Programa**:

**Entrada**: 5 hectáreas, Dosis de 2 L/ha.


**Cálculo**:Producto = 5 × 2 = 10 L
Agua = 10 × 500 = 5000 L

**Salida**:
10 L de producto y 5000 L de agua

```python
#Entrada de datos
hectarea=float(input("Ingrese el número de hectareas totales: "))
dosis = float(input("Ingrese la dosis requerida en litros por hectarea: "))

#Procesamiento
producto_total = hectarea * dosis # float
agua_total = producto_total * 500 # float

#Salida de datos
print("Litros de producto puro:", producto_total)
print("Litros de agua necesarios:", agua_total)#

```

#**Ejercicio 3**: Rendimiento de Pulpa en Frutas

**Planteamiento**:

Para una planta procesadora, es fundamental medir la eficiencia de la materia prima. El programa debe determinar cuánta pulpa aprovechable se obtiene tras retirar los residuos (cáscaras y semillas) para calcular la rentabilidad.

**Requerimientos del Problema**:

**Entradas**: Peso total de la fruta (kg) y Peso de los residuos (kg).

**Fórmulas**:

**Pulpa** = Peso Total − Residuos


**% Rendimiento** = (Pulpa / Peso Total) * 100

**Salida**:
Peso de la pulpa obtenida y el porcentaje de rendimiento.

**Prueba del Programa**:

**Entrada**: Peso Total = 500 kg, Residuos = 150 kg.


**Cálculo**:
Pulpa = 500 − 150 = 350 kg
Rendimiento = (350 / 500) × 100 = 70%

**Salida**:
350 kg de pulpa (70%)


```python
#Entrada de datos
peso_total = float(input("Ingrese el peso total de la fruta en kilogramos: "))
peso_residuos = float(input("Ingrese el peso de los residuos en kilogramos: "))

#Procesamiento
pulpa = peso_total - peso_residuos # float
porcentaje_rendimiento = (pulpa / peso_total) * 100 # Se añade el 100 para completar la fórmula del porcentaje.

#Salida
print("El peso de la pulpa obtenida es:", pulpa, "kg") # Se corrige 'peso_pulpa' a 'pulpa'.
print("El porcentaje de rendimiento es:", porcentaje_rendimiento, "%") # str y float
```

#**Ejercicio 4: Dosificación de Mortero para Muros**

#**Ejercicio 5: Peso de una Viga de Acero**
**Planteamiento**:
Calcular el peso propio de una viga de sección rectangular multiplicando su volumen por la densidad del acero.

**Requerimientos del Problema**:

**Entradas**: Base (m), Altura (m) y Longitud (m).

**Constantes**: Densidad del acero = 7850 kg/m3.

**Fórmulas**: Peso = Base * Altura * Longitud * 7850.

**Salida**: Peso total en kg.

**Prueba del Programa**:

**Entrada**: Base = 0.3m, Altura = 0.5m, Largo = 6m.

**Cálculo**: 0.3 * 0.5 * 6 * 7850 = 7065.

**Salida**: 7065 kg.

```python
# Entrada de datos
base = float(input("Ingrese la base de la viga (m): "))  # float
altura = float(input("Ingrese la altura de la viga (m): "))  # float
longitud = float(input("Ingrese la longitud de la viga (m): "))  # float

# Procesamiento
peso = base * altura * longitud * 7850  # float

# Salida de resultados
print("El peso de la viga es:", peso, "kg")  # str y float
```

#**Ejercicio 6: Pendiente de una Rampa de Acceso**
**Planteamiento**:
Determinar el porcentaje de inclinación de una rampa basándose en la altura y la distancia horizontal.

**Requerimientos del Problema**:

**Entradas**: Altura a vencer (m) y Longitud horizontal (m).

**Fórmulas**: %Pendiente = (Altura / Longitud) * 100.

**Salida**: Porcentaje de pendiente.

**Prueba del Programa**:

**Entrada**: Altura = 0.6m, Longitud = 5m.

**Cálculo**: (0.6 / 5) * 100 = 12.

**Salida**: 12% de pendiente.

```python
# Entrada de datos
altura_vencer = float(input("Ingrese la altura a vencer (m): "))  # float
longitud_horizontal = float(input("Ingrese la longitud horizontal (m): "))  # float

# Procesamiento
porcentaje_pendiente = (altura_vencer / longitud_horizontal) * 100  # float

# Salida de resultados
print("La pendiente de la rampa es:", porcentaje_pendiente, "%")  # str y float
```
